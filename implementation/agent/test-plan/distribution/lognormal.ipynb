```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the lognormal distribution
mu = 4.5  # Controls where the peak is (in log-space)
sigma = 0.6  # Controls the skew / burstiness

n_users = 100_000
arrival_times = np.random.lognormal(mean=mu, sigma=sigma, size=n_users)

# Convert to a limited window (e.g., 0-600 seconds)
arrival_times = arrival_times[arrival_times <= 600]

# Plotting histogram to visualize and get the histogram values
counts, bin_edges = np.histogram(arrival_times, bins=20)
plt.hist(arrival_times, bins=20, color='skyblue')
plt.title("Simulated Arrival Times of Users")
plt.xlabel("Seconds since ticket drop")
plt.ylabel("User count")
plt.show()

import json
arr = []

for i in range(len(counts)):
    arr.append({
        "target": int(counts[i]),
        "duration": "30s"
    })
    print(f"Bin {i+1}: {bin_edges[i]:.2f} - {bin_edges[i+1]:.2f} seconds: {counts[i]} users")

print(json.dumps(arr))
```
